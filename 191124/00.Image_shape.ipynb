# 준비 실습. Image 살펴보기

## 1. package 가져오기


```python
from PIL import Image
```

## 2. 이미지 가져오기


```python
image_gray = Image.open('01image.png')
image_color = Image.open('02image.png')

image_gray.show() # 로컬에서는 작은 창에 이미지가 뜸

print(image_gray.size) # 픽셀
print(image_gray)

print(image_color.size)
print(image_color)
```

## 3. 이미지 array로 변환하기


```python
import numpy as np

image_array = np.array(image_gray)
image_array_color = np.array(image_color)

print(image_array.shape) # (28, 28, 1) : 1은 dapth 또는 chaanel, 흑백을 의미, 생략 가능
print(image_array_color.shape) # 3은 컬러를 의미
```

```python
reshaped_1d = image_array.reshape(28*28) # 1차원 벡터, MLP에 넣을 수 있는 input.dim 형식
print(reshaped_1d)

reshaped_3d = image_array.reshape(28,28,1) # 3차원, CNN에 넣을 수 있는 input.shape 형식
print(reshaped_3d)
```

## 4. 이미지 살펴보기


```python
print(image_array)
```

```python
print(image_array_color)
```
