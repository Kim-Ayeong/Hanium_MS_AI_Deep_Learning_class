### 2019.11.23. 딥-러닝 과정 Mulit Layer Perceptron(MLP)

## 두번째 실습. Simple Keras 모델 생성/학습 - XOR Function

```python
# 1. Numpy 가져오기
import numpy as np

print(np.__version__)
```

```python
# 2. 입력/출력 데이터 만들기
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

print(X)
print(y)
```

```python
# 3. Keras 패키지 가져오기
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Activation

print(keras.__version__)
```

```python
# 4. MLP 모델 생성
model = Sequential()

model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

print(model.summary())
```

```python
# 5. Compile - Optimizer, Loss function 설정
model.compile(loss='binary_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
# 6. 학습시키기
batch_size = 1
epochs = 1000

model.fit(X, y,
          epochs=epochs,
          batch_size=batch_size,
          shuffle=True,
          verbose=0)
```

```python
# 7. 모델 테스트하기
predict = model.predict(np.array([[0,1],]))

print(predict)
```
